```python
import numpy as np
import pandas as pd
from Fuentes.ClassNeuronaLineal import *

# Datos de la tabla
horas_entrenamiento = np.array([2, 3, 4, 4, 5, 6, 6, 7, 7, 8, 10, 10])
tiempo_promedio = np.array([30.5, 29.3, 26.7, 26.0, 25.5, 24.7, 23.5, 23.0, 22.5, 21.0, 21.0, 20.5])

data = pd.DataFrame({
  'horas_entrenamiento': horas_entrenamiento,
  'tiempo_promedio': tiempo_promedio
})

configs = [
	{
		"w": -0.9013,
		"b": 29.5201
	}
	,{
		"w": -1.1301,
		"b": 31.2102
	},
 	{
		"w": -0.7080,
		"b": 30.0401
	}	
]

ALPHA = 0.01
MAX_ITE = 1000
COTA = 10e-6

nl_lin = NeuronaLineal(alpha=ALPHA, n_iter=MAX_ITE, cotaE=COTA, draw=0, title=["Entrenamiento [hs]", "Tiempo de estudio [min]"])

X = data['horas_entrenamiento'].values.reshape(-1, 1)
y = data['tiempo_promedio'].values

mejor_mse = float('inf')
mejor_w = None
mejor_b = None

for config in configs:
    nl_lin.w_ = np.array([config['w']])
    nl_lin.b_ = config['b']
    Y_pred = nl_lin.predict(X)
    mse = np.mean((y - Y_pred) ** 2)
    print(f"w: {nl_lin.w_[0]}, b: {nl_lin.b_}, ECM: {mse}")
    if mse < mejor_mse:
        mejor_mse = mse
        mejor_w = nl_lin.w_.copy()
        mejor_b = nl_lin.b_

# Usar la mejor configuración para la predicción final
nl_lin.w_ = mejor_w
nl_lin.b_ = mejor_b
hs_to_predict = np.array([[6]])
prediccion = nl_lin.predict(hs_to_predict)
print("Un corredor que entrena 6 horas semanales tardaría en promedio usando el mejor:", prediccion[0], "minutos")
```
